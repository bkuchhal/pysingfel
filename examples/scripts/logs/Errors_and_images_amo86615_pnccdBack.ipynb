```python
import numpy as np
import psana
import matplotlib.pyplot as plt
```

```python
errors = np.load('Errors_amo86615_pnccdBack.npy')
error_list = []
for i in range(190414):
    error_list.append(errors.item().get(i))
```

```python
plt.figure(figsize=(20, 8))
plt.plot(np.arange(0, len(error_list), 1), error_list)
plt.xlabel('Events')
plt.ylabel('Errors per pixel')
plt.title('Average error for every event')
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.plot(np.arange(0, len(error_list), 1), np.sort(error_list))
plt.ylabel('Error per pixel')
plt.title('Sorted Errors')
plt.show()
```

```python
dsname = 'exp=amo86615:run=197:idx'
ds = psana.DataSource(dsname)
run = ds.runs().next()
times = run.times()
env = ds.env()
numEvents = len(times)
detname = 'pnccdBack'
det = psana.Detector(detname, env)
```

```python
min_sort_list = np.sort(error_list)[3000:3010]
for i in range(len(error_list)):
    if error_list[i] in min_sort_list:
        print 'nevent=', i, 'error=', error_list[i]
        evt = run.event(times[i])
        img = det.image(evt)
        plt.figure(figsize=(10, 8))
        plt.imshow(img)
        plt.show()
```

```python
max_sort_list = np.sort(error_list)[-10:-1]
for i in range(len(error_list)):
    if error_list[i] in max_sort_list:
        print 'nevent=', i, 'error=', error_list[i]
        evt = run.event(times[i])
        img = det.image(evt)
        plt.figure(figsize=(10, 8))
        plt.imshow(img)
        plt.show()
```
